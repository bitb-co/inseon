# 2. 음성
 
 
---
*   음성인식: 컴퓨터가 음성정보를 문자데이터로 바꾸는 기술.
보편적으로 쓰이는 알고리즘은 HMM이다. HMM은 음성을 통계적으로 모델링하여 음향모델로 구성한뒤 다음에 입력하는 음성과 음향모델과 비교한다. 현재 가장 효율적인 방법이며 다양한 음을 어느정도 정확하게 인식할 수 있는 방법이다.
몇년전까지만해도 딱히 유망하지않았으나 스마트폰에 이식되면서 현재는 많은 연구가 이루어지고 있다.
 *   시리(siri)
     *   애플에서 제공하는 음성인식서비스를 기반으로 한 지능형 개인 비서 
     *   음성을 입력하면 자연어 처리를 한 뒤 그에 맞는 답변을 하거나 웹 서핑,
     애플리케이션 실행 등 다양한 업무를 처리한다.
     *  학습인식 기능이 있어서 사용자의 음성을 학습하여 인식률을 높일 수 있다.
     * 기존의 음성인식서비스는 지정된 사용어를 입력해야 작동이 가능했으나  문맥을 인식할 수 있는 자연어 처리덕분에 다양한 표현을 입력해서 실행시킬 수 있다.

*   이미지인식: 이미지나 영상에 비치는 문자와 얼굴 등 물체나 특징을 감지하는 기술. 이미지를 여러패턴으로 분류해서 패턴형데이터를 학습하여 새로운 이미지가 주어질 때 이것이 무엇인지 정확하게 알아낸다. 딥러닝이나 신경망 네트워크로 학습하는데 주로 쓰이는 신경망은 이미지의 형상을 그대로 보존할 수 있는 합성곱 신경망(CNN)이다.
   * 구글 렌즈
     *   구글이 개발한 인공지능인 '구글 어시스턴트'가 들어간 스마트 카메라 앱
     * 스마트폰에 내장되어있는 카메라로 사물을 촬영하면 어떤 사물인지 확인하고 
     필요한 정보를 찾아준다.

     *   지원하는 기능은 사진에서 텍스트를 추출하여 번역하는 스마트 텍스트,촬영된 사물과 비슷한 사물을 찾아주는 스타일 매치,다른 기종의 스마트폰에도 지원가능한 서드파티 카메라 등 다양하다.

* 자율주행차: 운전자가 브레이크,핸들,가속 페달 등을 직접 조종하지않아도 자동으로 상황을 파악해 스스로 운전하는 자동차이다. 자율주행차는 3d카메라,음파장비, 레이더, 라이다같은 여러가지 센서로 주변을 파악하고 인공지능이 판단하여 기계장치들을 제어한다.그렇기때문에 고속도로 주행지원시스템, 후측방 경보 시스템, 자동 긴급 제동 시스템, 차선이탈경보시스템 등 다양한 시스템들이 필요하다.  
   *  웨이모(waymo)
    *   구글의 자회사인 알파벳에서 만든 자율주행서비스
    *   '카크래프트'라는 자율주행시뮬레이터로 가상공간에서 시험해서 누적된 데이터를 심층신경망으로 갛화학습하여 실제 주행에 적용하였다.
    * 강화학습은 현재상황을 관찰하여 많은 보상이 있는 최선의 전략을 만드는 기계학습으로 다른 기계학습과 다르게 많은 데이터를 필요로 하지않는다.
    * '웨이모 드라이버'라는 자율주행 센서를 공개하였는데 이 센서는 라이다,레이더,카메라로 이루어져있다. 기존의 제품들보다 설계와 제조가 단순하여서 가격이 저렴하고 성능도 매우 뛰어나다.
